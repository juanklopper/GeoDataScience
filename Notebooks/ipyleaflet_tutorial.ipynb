```python
from ipywidgets import HTML
from ipyleaflet import Map, basemaps, basemap_to_tiles, Marker, Popup, Icon
```

```python
# Create a list of the longitude and latitude coordinates of Washington, D.C.
dc_coords = [38.9072, -77.0369]

# Set a zoom level
zoom_level = 12

# Create a map centered on Washington, D.C.
m = Map(center=dc_coords, zoom=zoom_level)

# Add a basemap
m.add_layer(basemaps.OpenStreetMap.Mapnik)

m
```

```python
Map(basemap=basemaps.Esri.WorldStreetMap, center=dc_coords, zoom=zoom_level)
```

```python
Map(basemap=basemaps.Esri.WorldImagery, center=dc_coords, zoom=zoom_level)
```

```python
Map(basemap=basemaps.NASAGIBS.ModisTerraTrueColorCR, center=dc_coords, zoom=zoom_level)
```

```python
m = Map(center=dc_coords, zoom=zoom_level)

dark_matter_layer = basemap_to_tiles(basemaps.CartoDB.Positron)
m.add(dark_matter_layer)
m
```

```python
marker = Marker(location=dc_coords, draggable=False, title="Washington, D.C.", z_index_offset=1000)
m.add(marker);

m
```

```python
center = [38.92, -77.02]
#dc_coords = [38.9072, -77.0369]

m = Map(center=dc_coords, zoom=zoom_level, close_popup_on_click=False)

marker = Marker(location=dc_coords)
m.add(marker)

message1 = HTML()
message2 = HTML()
message1.value = "Try clicking the marker!"
message2.value = "Hello <b>World</b>"
message2.placeholder = ""
message2.description = ""

# Popup with a given location on the map:
popup = Popup(
    location=center,
    child=message1,
    close_button=False,
    auto_close=False,
    close_on_escape_key=False
)
m.add(popup)

# Popup associated to a layer
marker.popup = message2

m
```
